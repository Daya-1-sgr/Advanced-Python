```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/'
response=requests.get(url)
print(response)
```

```python
cov19=BeautifulSoup(response.content,'html.parser')
print(cov19)
```

```python
cov19.select('td')[3].text
```

```python
r1=cov19.find('td').text
print(r1)
```

```python
datafull=cov19.find_all('td')
print(datafull)
print(len(datafull))
```

```python
Country=[]
Cases=[]
Deaths=[]
Region=[]
for i in range(0,len(datafull),4):
    Country.append(cov19.select('td')[i].text)
    Cases.append(cov19.select('td')[i+1].text)
    Deaths.append(cov19.select('td')[i+2].text)
    Region.append(cov19.select('td')[i+3].text)
    
```

```python
case1=[i.replace(',','') for i in Cases]
death1=[i.replace(',','') for i in Deaths]
case1
```

```python
import pandas as pd
df=pd.DataFrame({'Country':Country,'Cases':case1,'Deaths':death1,'Region':Region})
df.head()
```

```python
df.dtypes
```

```python
df['Cases']=df['Cases'].astype('int')
df['Deaths']=df['Deaths'].astype('int')
df.dtypes
```
